## Argumentos y par치metros

En la definici칩n de una funci칩n, los valores que se reciben se denominan **par치metros**, pero durante la llamada los valores que se env칤an se denominan **argumentos**.

Muchas veces estos t칠rminos se usan de manera intercambiable, pero es importante conocer la diferencia.

Un par치metro es una variable en la definici칩n de una funci칩n. Es s칩lamente un s칤mbolo y por eso no tiene un valor concreto. 

Un argumento en cambio es un valor que se le pasa a la funci칩n en la invocaci칩n. Poro decirlo de este modo, un argumento llena el campo que tiene reservado para 칠l un par치metro. 

```
def sumar(x,y):     # x, y son par치metros
  return x+y

sumar(5+8)          # 5, 8 son argumentos
```
Veamos un ejemplo de una funci칩n parametrizable.

Ejemplo de una funci칩n para calcular el vol칰men de un cilindro.

## Alcance o 치mbito de las variables

En `Python` (y en general tambi칠n en otros lenguajes de programaci칩n), las variables se encuentran confinadas dentro de un espacio. El t칠rmino en ingl칠s es _scope_ y en castellano se puede traducir como 치mbito o alcance. 

Hay dos tipos de alcance, **global** y **local**. 

Una variable **global** se puede invocar en cualqueir parte del programa. Es decir, se puede invocar incluso desde adentro de muchas funciones. 

````python
varGlobal = "Esta es una variable global" 

def funcion1():     # declaraci칩n
    print(varGlobal)

funcion1()          # invocaci칩n

def funcion2():     # declaraci칩n
    print(varGlobal)

funcion2()          # invocaci칩n


print(varGlobal)            # uso de la variable global  
````

Una variable **local** 칰nicamente existe dentro del cuerpo de la funci칩n donde esta se declara. 



````python
def funcion():
  varLocal = "Esta es una variable local"      
  print(varLocal)

funcion()
````

Si se invoca una variable local fuera de la funci칩n donde esta se declar칩 el sistema genera un error.

````python
print(varLocal)
````

Se puede crear una variable **local** con el mismo nombre de una variable **global**, pero `Python; lo ver치 como 2 variables distintas.

````python
miVariable = "Esta es una variable global" 

def funcion():
  miVariable = "Esta es una variable local"      
  print(miVariable)

funcion()             # utilizar치 la variable local 

print(miVariable)     # utilizar치 la variable global
````
